```python
###### Computational Cost #######


%matplotlib notebook

import matplotlib.pyplot as plt
import numpy as np
from matplotlib.animation import FuncAnimation
import psutil
import collections

plt.rcParams['grid.alpha'] = 1
plt.rcParams['grid.color'] = "#949292"
plt.rcParams['grid.linestyle'] = "--"
plt.rcParams['font.sans-serif'] = "Times New Roman"
plt.rcParams['font.family'] = "sans-serif"
legend_properties = {'weight':'bold'}

y = []
frame_len = 200
# p = psutil.Process()
pr = psutil.Process(29863)
# function to update the data
def my_function(i):
    
    # get data
#     cpu.popleft()

    cpu.append(pr.cpu_percent(1))  
#     ram.popleft()
    ram.append((pr.memory_info_ex()[0])/1e6)
    
    
    # clear axis
    ax.cla()
    ax1.cla()
    
    # plot cpu
    ax.plot(cpu,'r', marker='.')
    ax.scatter(len(cpu)-1, cpu[-1])
#     ax.text(len(cpu)-4, cpu[-1]+2, "{}%".format(cpu[-1]))
    ax.text(len(cpu)-1, cpu[-1]+2, "{}%".format(cpu[-1]))
    ax.set_ylim(0,100)
#     ax.set_xlabel('Time')
    ax.set_ylabel('CPU Usage (%)')
    ax.patch.set_facecolor('#ebebeb')
    ax.patch.set_alpha(0.2)
    ax.grid()
#     ax.set_yscale('log')
#     ax.set_title('')
    
    # plot memory
    ax1.plot(ram,'c', marker='.')
    ax1.scatter(len(ram)-1, ram[-1])
#     ax1.text(len(ram)-8, ram[-1]-30, "{:.3f} MB".format(ram[-1]))
    ax1.text(len(ram)-1, ram[-1]+2, "{:.3f} MB".format(ram[-1]))
#     ax1.set_ylim(0,200)
    ax1.set_xlabel('Time')
    ax1.set_ylabel('RAM Usage (MB)')
    ax1.patch.set_facecolor('#ebebeb')
    ax1.patch.set_alpha(0.2)
    ax1.grid()
#     ax1.legend()
#     ax1.set_yscale('log')
#     ax1.set_title('RAM (GB)')

# start collections with zeros
cpu = collections.deque(np.zeros(10))
ram = collections.deque(np.zeros(10))
# define and adjust figure
fig = plt.figure(facecolor='w')
fig.tight_layout()


ax = plt.subplot(211)
ax1 = plt.subplot(212)
ax.set_facecolor('w')
ax1.set_facecolor('w')


# animate
ani = FuncAnimation(fig, my_function, interval=1000)

plt.show()

```
